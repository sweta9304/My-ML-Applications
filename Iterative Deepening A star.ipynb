```python
#Code Block : Write function to design the Transition Model/Successor function. Ideally this would be called while search algorithms are implemented
# while identiying transition we need to calculate h(n) for each position 
# Algorithm 2 - Local Search (Hill Climbing Algorithm)

def successor_position(state):  
    
    coordinates = list(state)
    
    x = coordinates[0]
    y = coordinates[1]
    
    
    if (x,y) not in lstInitialStates:
        lstInitialStates.append((x,y))
    
    dctNextPos = {}
    
    if x > 0 and x <= rowsCount:# this means within valid row limit
        if y > 0 and y <= columnsCount: # this means within valid column limit
            # next possible position could be in 4 directions
            # but based on current position we can clearly see next possible positions
            # so from the current position next probable positions will be where no walls
            # check for directions with 0
            dirWithNoWalls = {new_key:new_value for (new_key, new_value) in maze.get((x,y)).items() if new_value == 0}
            #print(dirWithNoWalls)
            for direction in dirWithNoWalls:
                if str(direction) == 'E':
                    if((x,y+1) in maze):                        
                        if((x,y+1) not in lstInitialStates):                            
                            dctNextPos[(x,y+1)] = abs(goalRow - x) + abs(goalCol - (y+1)) + 3 + 5                            
                elif str(direction) == 'W':
                     if((x,y-1) in maze):                        
                        if((x,y-1) not in lstInitialStates):
                            dctNextPos[(x,y-1)] = abs(goalRow - x) + abs(goalCol - (y-1)) + 3                            
                elif str(direction) == 'N':
                     if((x-1,y) in maze):                        
                        if((x-1,y) not in lstInitialStates):
                            dctNextPos[(x-1,y)] = abs(goalRow - (x-1)) + abs(goalCol - y) + 3                            
                elif str(direction) == 'S':
                     if((x+1,y) in maze):                        
                        if((x+1,y) not in lstInitialStates):   
                            dctNextPos[(x+1,y)] = abs(goalRow - (x+1)) + abs(goalCol - y) + 3                           
            
            # when we have two items choose best item from the dictonary of positions
            bestPosition = min(dctNextPos, key=dctNextPos.get)
            
            #print(bestPosition)
            
            dctBestNextPos = {new_key:new_value for (new_key, new_value) in dctNextPos.items() if new_key == bestPosition}
            
            return dctBestNextPos

#successor_position((6,2))
```

### 2.	Definition of Algorithm 1 (ITERATIVE DEEPENING A-STAR SEARCH)

```python
#Code Block : Function for algorithm Iterative Deepning A-star Search
import timeit
import numpy as np
import sys

# Below are driver functions to implement IDA-star algorithm using Graph Data Structure and its adjacency List
class Graph:
    
    # Contains graph info in the form of adjacency matrix
    def __init__(self, adjacency_list):
        self.adjacency_list = adjacency_list

    def getNeighbors(self, v):
        return self.adjacency_list[v]

    # heuristic data for each node with base cost 3 and with penalty 5 for East direction(Explained in theory document)
    def heuristic(self, n):
        H = {
            'A': 60,
            'B': 54,
            'C': 6,
            'G': 0,
            'D': 68,
            'F': 80,
            'H': 86,   
            'E': 77,
            'I': 85,
            'K': 80,
            'J': 77,
            'L': 69,
            'M': 57,
            'N': 49,
            'O': 37,
            'P': 29,
            'Q': 17,
            'R': 9
        }
        if n not in H.keys():
            return -1
        else:
            return H[n]
    
    def is_valid_maze_position(self, node):
        return self.heuristic(node)
    
    def aStarAlgorithmRecur(self, start_node, stop_node, threshold):
        # open_list having nodes which have been visited, but neighbours are still unvisited
        # closed_list is a list of nodes which have been visited along with its neighbours
        o_len = 0
        c_len = 0
        
        if self.is_valid_maze_position(start_node) == -1 or self.is_valid_maze_position(stop_node) == -1:
            print("Invalid Start Node or Goal Node\n")
            return -1
        
        open_list = set([start_node])
        type(open_list)
        closed_list = set([])

        # g dict contains current distances from start_node to all other nodes
        g = {}

        g[start_node] = 0
        
        # parents contains an adjacency map of all nodes
        parents = {}
        parents[start_node] = start_node

        while len(open_list) > 0:
            n = None

            # find a node with the lowest value of f() - evaluation function
            for v in open_list:
                if n == None or g[v] + self.heuristic(v) < g[n] + self.heuristic(n):
                    n = v;

            if n == None:
                print('Path does not exist!')
                return None

            # if level or threshold is less than f(n), return and restart with new threshold
            if threshold < g[n]:
                return g[n]
            else:
                # if the current node is the stop_node
                # then we begin reconstructing the path from it to the start_node
                if n == stop_node:
                    print("\nGoal Node is found!!")
                    reconst_path = []

                    # Construct path till starting node
                    while parents[n] != n:
                        reconst_path.append(n)
                        n = parents[n]

                    reconst_path.append(start_node)
                    reconst_path.reverse()

                    print('Path found: {}'.format(reconst_path))
                    cost_list = []
                    for n in reconst_path:
                        cost_list.append(g[n])
                    print('Path Cost: {}'.format(cost_list))
                    print("Solution found with heuristic-upperbound: {}\n".format(threshold))
                    print('Space Complexity: Open list max depth: {}'.format(o_len))
                    print('Space Complexity: Closed list max depth: {}'.format(c_len))
                    return reconst_path


                # for all neighbors of the current node, process them
                for (m, weight) in self.getNeighbors(n):

                    # if the current node isn't in both open_list and closed_list
                    # add it to open_list and note n as it's parent
                    
                    if m not in open_list and m not in closed_list:
                        open_list.add(m)
                        
                        # save max depth of open list
                        if (o_len < len(open_list)):
                            o_len = len(open_list)
                            
                        # Assign parents and f(n) to the nodes
                        parents[m] = n
                        g[m] = g[n] + weight

                    # otherwise, check if it's quicker to first visit n, then m
                    # and if it is, update parent data and g data
                    # and if the node was in the closed_list, move it to open_list
                    else:
                        if g[m] > g[n] + weight:
                            g[m] = g[n] + weight
                            parents[m] = n

                            if m in closed_list:
                                closed_list.remove(m)
                                open_list.add(m)

                # remove n from the open_list, and add it to closed_list
                # because all of his neighbors were inspected
                open_list.remove(n)
                closed_list.add(n)
                
                # save max depth of closed list
                if (c_len < len(closed_list)):
                    c_len = len(closed_list)

        print('Path does not exist!')
        return 0

    def space_complexity(self,adjacency_list):
        space_comp = 0
        for keys in adjacency_list.keys():
            # Calculate space for adjacency dictionary
            space_comp += len(keys) + len(adjacency_list[keys])
            
        return space_comp
```

### 3.	Definition of Algorithm 2 (HILL CLIMBING SEARCH)

```python
#Code Block : Function for algorithm 2 implementation
# so base on requirement of hill climbing algorithm lets find initial state cost and check 
# with neighbours to go with least state, with no memory saving
def is_better(state1, state2):
    #print("current state : ", state2)
    #print("future state : ", state1)
    if state1 > state2:
        return True
    else:
        return False

def hill_climbing(x,y):
    currentState = {startPosition: initialHeurVal}
    #print("Current",currentState)
    goalState = goalPosition
    path = {(x,y):initialHeurVal}
    print("path is :",path)
    while not goal_check(list(currentState.keys())[0]):
        found = False        
        newState = successor_position(list(currentState.keys())[0])
        #print("New State",newState)
        if is_better(list(newState.values())[0], list(currentState.values())[0]): 
            path.update(newState)
            print("path updated",newState)
            currentState = newState
            #print("updated currentstate",currentState)
            found = True            
        if found:
            continue
        return path, "Goal Not reached as local maxima reached"
    return path, "Goal Reached"
```

```python
# Invoke algorithm 1 (Should Print the solution, path, cost etc., (As mentioned in the problem))
# Caller block for IDA-star algorithm

graph1 = Graph(adjacency_list)

# Calling Iterative deepening A-Star algorithm recursively with increasing thresholds starting threshold with a value 1

threshold = 1
distance=0

print("Initial Level/Threshold for IDAstar Algo:{} \n".format(threshold))

# Using timeit function to determine complexity at the end of algorithm
start = timeit.default_timer()

# Call A-star Algo with increasing weights until Goal is found
while True:

    print("Iteration with threshold: " + str(threshold))
    distance = graph1.aStarAlgorithmRecur('A', 'G',threshold)
    
    # Goal Node is found, exit the loop
    if type(distance) == list:
        break
    # Path does not exist, so exit the loop
    if distance == 0:
        break
    # Invalid Start or Goal node given in the input
    if distance == -1:
        break
    
    if type(distance) != list and distance == float("inf"):
        # Node not found and no more nodes to visit
        print("Not Found the node we're looking for!")
    else:
    # if it hasn't found the node, it returns the (positive) next-bigger threshold
        threshold = distance

stop = timeit.default_timer()
```
